```python
# Creating a Sample Cloud Infrastructure Dataset


cloud_data = {
    "Server": [
        "Server-01", "Server-02", "Server-03", "Server-04", "Server-05",
        "Server-06", "Server-07", "Server-08", "Server-09", "Server-10"
    ],
    "CPU_Usage (%)": [35, 67, 91, 54, 82, 43, 76, 59, 88, 47],
    "Memory_Usage (%)": [48, 72, 88, 63, 76, 55, 81, 66, 90, 58],
    "Disk_Usage (%)": [52, 61, 79, 57, 73, 47, 69, 60, 82, 54],
    "Network_Latency (ms)": [12, 20, 35, 18, 28, 15, 26, 19, 33, 16],
    "Temperature (°C)": [42, 57, 74, 49, 68, 45, 64, 53, 72, 47]
}

df = pd.DataFrame(cloud_data)

print("Cloud Infrastructure Monitoring Dataset")
display(df)
```

## Numerical Computing with NumPy

NumPy provides efficient numerical operations on multidimensional arrays. In this section, selected cloud infrastructure metrics are converted into NumPy arrays to perform statistical analysis, helping identify the overall utilization of system resources.

```python
# Extract CPU usage values as a NumPy array

cpu_usage = df["CPU_Usage (%)"].to_numpy()

print("CPU Usage Array:")
print(cpu_usage)

print("\nAverage CPU Usage      :", np.mean(cpu_usage))
print("Maximum CPU Usage      :", np.max(cpu_usage))
print("Minimum CPU Usage      :", np.min(cpu_usage))
print("Standard Deviation     :", round(np.std(cpu_usage), 2))
```

```python
# ==========================================
# Statistical Summary
# ==========================================

print("Statistical Summary of Numerical Features")

display(df.describe())
```

```python
# ==========================================
# Missing Value Analysis
# ==========================================

print("Missing Values in Each Column")

display(df.isnull().sum())
```

## Visualizing Cloud Infrastructure Metrics

Data visualization is an essential step in machine learning as it helps identify patterns, trends, and anomalies within the data. In this section, various plots are generated to visualize the cloud infrastructure metrics collected from the sample monitoring dataset.

```python
plt.figure(figsize=(10,5), facecolor='black')

ax = plt.gca()
ax.set_facecolor('black')

plt.plot(
    df["Server"],
    df["CPU_Usage (%)"],
    marker='o',
    linewidth=2,
    color='darkblue'
)

plt.title("CPU Utilization Across Cloud Servers", color='white')
plt.xlabel("Server", color='white')
plt.ylabel("CPU Usage (%)", color='white')

plt.grid(True, color='gray', linestyle='--', alpha=0.7)

plt.xticks(rotation=45, color='white')
plt.yticks(color='white')

plt.show()
```

```python
plt.figure(figsize=(10,5), facecolor='black')

ax = plt.gca()
ax.set_facecolor('black')

plt.bar(
    df["Server"],
    df["Memory_Usage (%)"],
    color='darkblue'
)

plt.title("Memory Utilization Across Cloud Servers", color='white')
plt.xlabel("Server", color='white')
plt.ylabel("Memory Usage (%)", color='white')

plt.xticks(rotation=45, color='white')
plt.yticks(color='white')

plt.show()
```

```python
plt.figure(figsize=(8,5), facecolor='black')

ax = plt.gca()
ax.set_facecolor('black')

box_plot = plt.boxplot(
    [
        df["CPU_Usage (%)"],
        df["Memory_Usage (%)"],
        df["Disk_Usage (%)"]
    ],
    boxprops=dict(color='darkblue'),
    medianprops=dict(color='white'),
    whiskerprops=dict(color='darkblue'),
    capprops=dict(color='darkblue'),
    flierprops=dict(markeredgecolor='darkblue')
)

plt.xticks(
    [1,2,3],
    ["CPU","Memory","Disk"],
    color='white'
)

plt.ylabel("Utilization (%)", color='white')

plt.title("Distribution of Resource Utilization", color='white')

plt.yticks(color='white')

plt.show()
```

```python
# ==========================================
# Correlation Matrix
# ==========================================

correlation = df.drop(columns=["Server"]).corr()

display(correlation)
```

## Loading a Sample Dataset Using Scikit-learn

Scikit-learn provides several built-in datasets that are commonly used for learning and validating machine learning workflows. In this experiment, the Digits dataset is loaded to verify that the machine learning environment is functioning correctly. The dataset consists of handwritten digit images represented as numerical features and is widely used for introductory machine learning and pattern recognition tasks.

```python
# ==========================================
# Load the Digits Dataset
# ==========================================

from sklearn.datasets import load_digits

digits = load_digits()

print("Digits dataset loaded successfully.")
```

```python
# ==========================================
# Convert to a Pandas DataFrame
# ==========================================

digits_df = pd.DataFrame(
    digits.data,
    columns=[f"Pixel_{i}" for i in range(digits.data.shape[1])]
)

digits_df["Digit"] = digits.target

display(digits_df.head())
```

```python
# ==========================================
# Dataset Information
# ==========================================

print("Dataset Shape:", digits_df.shape)

print("\nNumber of Features:", len(digits.feature_names))

print("\nTarget Classes:", digits.target_names)
```

```python
# ==========================================
# Statistical Summary
# ==========================================

display(digits_df.describe())
```

```python
# ==========================================
# Missing Value Analysis
# ==========================================

display(digits_df.isnull().sum())
```

## Visualizing a Sample Image

The Digits dataset contains grayscale images of handwritten digits represented as 8 × 8 pixel grids. Visualizing a sample image confirms that the dataset has been loaded correctly and provides an intuitive understanding of the data.

```python
# ==========================================
# Display a Sample Digit
# ==========================================

plt.figure(figsize=(4,4))

plt.imshow(digits.images[0], cmap="gray")

plt.title(f"Sample Digit: {digits.target[0]}")

plt.axis("off")

plt.show()
```

## Random Number Generation using NumPy

Random number generation plays an important role in machine learning. It is commonly used for tasks such as initializing model parameters, shuffling datasets, generating synthetic data, and creating reproducible experiments. By setting a random seed, the same sequence of random numbers can be generated every time the code is executed.

```python
# ==========================================
# Random Number Generation
# ==========================================

np.random.seed(42)

random_cpu_usage = np.random.randint(20, 100, size=10)

print("Randomly Generated CPU Utilization Values (%)")

print(random_cpu_usage)
```
